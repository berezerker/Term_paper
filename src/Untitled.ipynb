```python
def get_path_names(folder):
  images = []
  for path, subdirs, files in os.walk(folder):  
    for name in files: 
      images.append(os.path.join(path, name))
  return images
```

```python
path = "test/rain"
for file in os.listdir(path):
    print(file)
    img = cv2.imread(os.path.join("test/rain", file))
    RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_filtered = cv2.GaussianBlur(RGB_img, (7, 7), 0)
    dest_name = os.path.join("test/rain_blur", file)
    im = Image.fromarray(img_filtered)
    im.save(dest_name)
```
