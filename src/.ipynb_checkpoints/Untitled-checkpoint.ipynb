```python
train_files= []
train_folders = os.listdir('demoir/train/train')
for folder_num, folder in enumerate(train_folders):
    train_files = (os.listdir(os.path.join('demoir/train/train', folder)))
    train_files_gt = train_files[::2]
    train_files_moir = train_files[1::2]
    for i in range(len(train_files_gt)):
        dest_gt = os.path.join('demoir/processed/processed_train/gt', str(folder_num)+ "_" + train_files_gt[i])
        dest_moir = os.path.join('demoir/processed/processed_train/moir', str(folder_num)+ "_" + train_files_moir[i])
        shutil.copyfile(os.path.join('demoir/train/train', folder, train_files_gt[i]), dest_gt)
        shutil.copyfile(os.path.join('demoir/train/train', folder, train_files_moir[i]), dest_moir)
        
```

```python
path = "test/rain"
for file in os.listdir(path):
    print(file)
    img = cv2.imread(os.path.join("test/rain", file))
    RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_filtered = cv2.GaussianBlur(RGB_img, (7, 7), 0)
    dest_name = os.path.join("test/rain_blur", file)
    im = Image.fromarray(img_filtered)
    im.save(dest_name)
```

```python
path = "test/demoir"
for file in os.listdir(path):
    print(file)
    img = cv2.imread(os.path.join(path, file))
    RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_filtered = cv2.GaussianBlur(RGB_img, (37, 37), 0)
    dest_name = os.path.join("test/demoir_blur", file)
    im = Image.fromarray(img_filtered)
    im.save(dest_name)
```
